```python
import torch, time, csv, gc, sys
sys.path.insert(0, '/kaggle/working/mvrepo/MambaVision-main')
sys.path.insert(0, '/kaggle/working/mvrepo/pom-main')
from mambavision.models.mamba_vision import mamba_vision_T, mamba_vision_S, mamba_vision_B

dev = 'cuda'
factories = {'T': mamba_vision_T, 'S': mamba_vision_S, 'B': mamba_vision_B}
resolutions = [224, 384, 512, 768]
variants = [('attn', False), ('pom', False), ('attn', True), ('pom', True)]
BATCH = 16          # drop to 8 if windowed variants OOM early
WARMUP, ITERS = 5, 20

def bench(model, x):
    model.eval()
    with torch.no_grad():
        for _ in range(WARMUP): model(x)
        torch.cuda.synchronize(); torch.cuda.reset_peak_memory_stats()
        t0 = time.perf_counter()
        for _ in range(ITERS): model(x)
        torch.cuda.synchronize()
        dt = (time.perf_counter() - t0) / ITERS
    return dt, torch.cuda.max_memory_allocated() / 1e6

rows = []
for size, fac in factories.items():
    for mt, g in variants:
        tag = f"{mt}_{'global' if g else 'win'}"
        nparams = None
        for res in resolutions:
            try:
                torch.cuda.empty_cache(); gc.collect()
                model = fac(pretrained=False, mixer_type=mt, global_mixer=g).to(dev)
                if nparams is None:
                    nparams = sum(p.numel() for p in model.parameters())
                x = torch.randn(BATCH, 3, res, res, device=dev)
                dt, peak = bench(model, x)
                ips = BATCH / dt
                rows.append([size, tag, res, round(ips,1), round(peak,1), nparams])
                print(f"{size:2s} {tag:11s} res={res:4d}  {ips:8.1f} img/s  peak={peak:9.1f} MB  params={nparams:,}")
                del model, x
            except RuntimeError as e:
                msg = 'OOM' if 'out of memory' in str(e).lower() else str(e)[:30]
                rows.append([size, tag, res, msg, '', nparams])
                print(f"{size:2s} {tag:11s} res={res:4d}  {msg}")
                try: del model, x
                except: pass
                torch.cuda.empty_cache(); gc.collect()

with open('/kaggle/working/benchmark_global.csv','w',newline='') as f:
    w = csv.writer(f)
    w.writerow(['model','variant','resolution','img_per_sec','peak_mem_MB','params'])
    w.writerows(rows)
print("\nSaved /kaggle/working/benchmark_global.csv")
```

```python
import torch, sys
sys.path.insert(0, '/kaggle/working/mvrepo/MambaVision-main')
sys.path.insert(0, '/kaggle/working/mvrepo/pom-main')
from mambavision.models.mamba_vision import mamba_vision_T

# Hook the LAST transformer block's mixer to capture the actual input sequence length it sees
def probe(global_mixer, res):
    m = mamba_vision_T(pretrained=False, mixer_type='attn', global_mixer=global_mixer).cuda().eval()
    seqs = []
    # find the last stage's blocks and hook their mixer forward
    last_stage = m.levels[-1]
    captured = {}
    def hook(mod, inp, out):
        x = inp[0]
        captured['shape'] = tuple(x.shape)
    h = last_stage.blocks[-1].mixer.register_forward_hook(hook)
    x = torch.randn(1, 3, res, res).cuda()
    with torch.no_grad():
        m(x)
    h.remove()
    del m; torch.cuda.empty_cache()
    return captured.get('shape')

for res in [224, 512, 768]:
    sw = probe(False, res)
    sg = probe(True, res)
    print(f"res={res:4d}  windowed mixer input={sw}   global mixer input={sg}")
```

```python
import torch, time, csv, gc, sys
sys.path.insert(0, '/kaggle/working/mvrepo/MambaVision-main')
sys.path.insert(0, '/kaggle/working/mvrepo/pom-main')
from mambavision.models.mamba_vision import mamba_vision_T

dev = 'cuda'
resolutions = [768, 1024, 1280, 1536, 2048]
variants = [('attn', True), ('pom', True)]   # global only
BATCH = 4            # small, to give high-res a chance before OOM
WARMUP, ITERS = 3, 10

def last_stage_seqlen(model, res):
    """Capture the token count N the final-stage mixer actually sees."""
    cap = {}
    def hook(mod, inp, out): cap['N'] = inp[0].shape[1]
    h = model.levels[-1].blocks[-1].mixer.register_forward_hook(hook)
    with torch.no_grad():
        model(torch.randn(1, 3, res, res, device=dev))
    h.remove()
    return cap.get('N')

def bench(model, x):
    model.eval()
    with torch.no_grad():
        for _ in range(WARMUP): model(x)
        torch.cuda.synchronize(); torch.cuda.reset_peak_memory_stats()
        t0 = time.perf_counter()
        for _ in range(ITERS): model(x)
        torch.cuda.synchronize()
        dt = (time.perf_counter() - t0) / ITERS
    return dt, torch.cuda.max_memory_allocated() / 1e6

rows = []
for res in resolutions:
    for mt, g in variants:
        tag = f"{mt}_global"
        try:
            torch.cuda.empty_cache(); gc.collect()
            model = mamba_vision_T(pretrained=False, mixer_type=mt, global_mixer=g).to(dev)
            # sequence length first (cheap, batch=1)
            try:
                N = last_stage_seqlen(model, res)
            except RuntimeError:
                N = '?'
            torch.cuda.empty_cache()
            x = torch.randn(BATCH, 3, res, res, device=dev)
            dt, peak = bench(model, x)
            ips = BATCH / dt
            rows.append([tag, res, N, round(ips,2), round(peak,1)])
            print(f"{tag:11s} res={res:4d}  N={str(N):>5s}  {ips:7.2f} img/s  peak={peak:9.1f} MB")
            del model, x
        except RuntimeError as e:
            msg = 'OOM' if 'out of memory' in str(e).lower() else str(e)[:30]
            rows.append([tag, res, locals().get('N','?'), msg, ''])
            print(f"{tag:11s} res={res:4d}  N={str(locals().get('N','?')):>5s}  {msg}")
            for v in ('model','x'):
                try: del locals()[v]
                except: pass
            torch.cuda.empty_cache(); gc.collect()

with open('/kaggle/working/benchmark_extreme.csv','w',newline='') as f:
    w = csv.writer(f)
    w.writerow(['variant','resolution','seq_len_N','img_per_sec','peak_mem_MB'])
    w.writerows(rows)
print("\nSaved /kaggle/working/benchmark_extreme.csv")
```

```python
import torch, sys, gc
sys.path.insert(0, '/kaggle/working/mvrepo/MambaVision-main')
sys.path.insert(0, '/kaggle/working/mvrepo/pom-main')
from mambavision.models.mamba_vision import mamba_vision_T

dev='cuda'
for res in [1024, 1536, 2048]:
    for mt in ['attn','pom']:
        try:
            torch.cuda.empty_cache(); gc.collect()
            m = mamba_vision_T(pretrained=False, mixer_type=mt, global_mixer=True).to(dev).train()
            torch.cuda.reset_peak_memory_stats()
            x = torch.randn(2,3,res,res,device=dev,requires_grad=True)
            out = m(x); loss = out.float().pow(2).mean()
            loss.backward()
            torch.cuda.synchronize()
            peak = torch.cuda.max_memory_allocated()/1e6
            print(f"{mt:4s}_global res={res:4d}  TRAIN peak={peak:9.1f} MB")
            del m,x,out,loss
        except RuntimeError as e:
            msg='OOM' if 'out of memory' in str(e).lower() else str(e)[:30]
            print(f"{mt:4s}_global res={res:4d}  TRAIN {msg}")
            torch.cuda.empty_cache(); gc.collect()
```
